# Verification on Questions to Genie

### Question 1: Which country has the most athletes?
Genie answer: United States has the most athletes
The United States has the most athletes with 247,956 unique athletes, significantly ahead of France (193,703) and South Africa (94,684).

```sql
SELECT country_name, COUNT(DISTINCT athlete_id) AS num_athletes
FROM marathos.gold.dim_athlete
GROUP BY country_name
ORDER BY num_athletes DESC
LIMIT 3
```

### Question 2: What is the gender distribution of athletes?

```sql
SELECT
  gender,
  COUNT(DISTINCT athlete_id) AS num_athletes,
  ROUND(
    COUNT(DISTINCT athlete_id) * 100.0 / SUM(COUNT(DISTINCT athlete_id)) OVER (),
    2
  ) AS percentage
FROM
  marathos.gold.dim_athlete
GROUP BY
  gender
```

### Question 3: How many events were held in 2022?

```sql
SELECT COUNT(DISTINCT event_name) AS num_events
FROM marathos.gold.dim_event
WHERE event_year = 2022
```

### Question 4: Which marathon event has the most finishers?

```sql
SELECT event_name, MAX(number_finishers) AS finishers
FROM marathos.gold.dim_event
GROUP BY event_name
ORDER BY finishers DESC
LIMIT 1
```

### Question 5: What is the average speed by gender?

```sql
SELECT a.gender, ROUND(AVG(f.average_speed), 2) AS avg_speed
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_athlete a ON f.athlete_id = a.athlete_id
WHERE f.average_speed > 0
GROUP BY a.gender
```

### Question 6: What are the 10 most popular events by number of participants?

```sql
SELECT
    e.event_name,
    e.event_type,
    COUNT(DISTINCT r.athlete_id) AS unique_participants
FROM marathos.gold.fct_results r
JOIN marathos.gold.dim_event e USING (event_id)
GROUP BY e.event_name, e.event_type
ORDER BY unique_participants DESC
LIMIT 10
```
