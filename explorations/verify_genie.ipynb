### Question 1: Which country has the most athletes?
Genie: The United States has the most athletes with 233,184 unique athletes, significantly ahead of France in second place with 192,479 athletes

Conclution: Got the same result when I changed to mart_distance_events from mart_overview

```sql
SELECT country_name, COUNT(DISTINCT athlete_id) AS num_athletes
FROM marathos.gold.mart_distance_events
WHERE country_name IS NOT NULL
GROUP BY country_name
ORDER BY num_athletes DESC
LIMIT 5
```

### Question 2: What is the gender distribution of athletes?
Genie: Male athletes significantly outnumber female athletes
The athlete population is 76.3% male and 23.7% female, with 900,212 male athletes compared to 279,693 female athletes.

```sql
SELECT
  gender,
  COUNT(DISTINCT athlete_id) AS num_athletes,
  ROUND(
    COUNT(DISTINCT athlete_id) * 100.0 / SUM(COUNT(DISTINCT athlete_id)) OVER (),
    1
  ) AS percentage
FROM
  marathos.gold.mart_distance_events
WHERE gender IS NOT NULL
GROUP BY
  gender
ORDER BY num_athletes DESC
```

### Question 3: What year has most events?
Genie: 2019 had the most events with 2,072 events
2019 had the highest number of ultramarathon distance events with 2,072 distinct events, followed by 2017 with 1,783 events and 2018 with 1,749 event

```sql
SELECT event_year,
COUNT(DISTINCT event_name) AS num_events
FROM marathos.gold.mart_distance_events
WHERE event_year IS NOT NULL
GROUP BY event_year
ORDER BY num_events DESC
LIMIT 10

-- colund't unsort values (like I did on the one above)
```

### Question 4: Which marathon event has the most finishers?
Genie: 
- Comrades Marathon - Up Run (RSA) - 20,027 finishers
- Comrades Marathon - Down Run (RSA) - 16,484 finishers
- Two Oceans Marathon (RSA) - 12,096 finishers
- Two Oceans Marathon - 50km Split (RSA) - 11,967 finishers
- La SaintéLyon 72 km (FRA) - 5,787 finishers


```sql
SELECT event_name, MAX(number_finishers) AS finishers
FROM marathos.gold.mart_overview
GROUP BY event_name
ORDER BY finishers DESC
LIMIT 5
```

### Question 5: What is the average speed by gender?
Genie: Male athletes average 7.35 km/h vs. female athletes at 6.94 km/h
Male athletes have an average speed of 7.35 km/h compared to female athletes at 6.94 km/h, a difference of approximately 0.41 km/h or about 6% faster.

```sql
SELECT gender, ROUND(AVG(average_speed), 2) AS avg_speed
FROM marathos.gold.mart_distance_events
WHERE average_speed > 0
GROUP BY gender
```

### Question 6: What are the 10 most popular events by number of participants?
Genie: Comrades Marathon - Up Run is the most popular event with 2.88 billion total participants
The Comrades Marathon - Up Run (RSA) leads with 2.88 billion total participants, followed by Comrades Marathon - Down Run with 2.41 billion and Two Oceans Marathon with 2.20 billion participants.

```sql
SELECT
    event_name,
    COUNT(DISTINCT athlete_id) AS unique_participants
FROM marathos.gold.mart_overview
GROUP BY event_name, event_type
ORDER BY unique_participants DESC
LIMIT 10
```
