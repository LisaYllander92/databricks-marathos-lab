# Verification on Questions to Genie

### Question 1: Which country has the most athletes?
Genie: United States has the most athletes
The United States has the most athletes with 247,956 unique athletes, significantly ahead of France (193,703) and South Africa (94,684).

```sql
SELECT country_name, COUNT(DISTINCT athlete_id) AS num_athletes
FROM marathos.gold.mart_overview
GROUP BY country_name
ORDER BY num_athletes DESC
LIMIT 5
```

### Question 2: What is the gender distribution of athletes?
Genie: Male athletes significantly outnumber female athletes
The athlete population is 76.3% male and 23.7% female, with 902,471 male athletes compared to 279,747 female athletes

```sql
SELECT
  gender,
  COUNT(DISTINCT athlete_id) AS num_athletes,
  ROUND(
    COUNT(DISTINCT athlete_id) * 100.0 / SUM(COUNT(DISTINCT athlete_id)) OVER (),
    1
  ) AS percentage
FROM
  marathos.gold.mart_distance_events
WHERE gender IS NOT NULL
GROUP BY
  gender
ORDER BY num_athletes DESC
```

### Question 4: Which marathon event has the most finishers?
Genie: Comrades Marathon - Up Run (2000) had the most finishers with 20,027
The Comrades Marathon - Up Run held in South Africa on June 16, 2000 had the most finishers with 20,027


```sql
SELECT event_name, MAX(number_finishers) AS finishers
FROM marathos.gold.mart_overview
GROUP BY event_name
ORDER BY finishers DESC
LIMIT 5
```

### Question 5: What is the average speed by gender?
Genie: Male athletes are 5.7% faster on average than female athletes
Male athletes have an average speed of 7.99 km/h compared to 7.55 km/h for female athletes, representing a difference of approximately 0.43 km/h or 5.7%

```sql
SELECT gender, ROUND(AVG(average_speed), 2) AS avg_speed
FROM marathos.gold.mart_overview 
WHERE average_speed > 0
GROUP BY gender
```
